```python
from math import sin
import matplotlib.pyplot as plt

# Create the main figure with specified size
plt.figure(figsize=(16, 9))

x = list(range(-10, 11))

# Plot y = x^2 in the first subplot (1st position in 3 rows, 4 columns)
y = [i**2 for i in x]
plt.subplot(3, 4, 1)  # 3 rows, 4 columns, position 1
plt.plot(x, y, label="y = x^2")
plt.title('Graph of x**2')
plt.xlabel('x values')
plt.ylabel('y = x^2')
plt.grid(True)
plt.legend()

# Plot z = x^3 in the tenth subplot (10th position in 3 rows, 4 columns)
z = [i**3 for i in x]
plt.subplot(3, 4, 10)  # 3 rows, 4 columns, position 10
plt.plot(x, z, label="y = x^3", color='orange')
plt.title('Graph of x**3')
plt.xlabel('x values')
plt.ylabel('y = x^3')
plt.grid(True)
plt.legend()

# Plot w = sin(x) in the eleventh subplot (11th position in 3 rows, 4 columns)
w = [sin(i) for i in x]
plt.subplot(3, 4, 11)  # 3 rows, 4 columns, position 11
plt.plot(x, w, label="y = sin(x)", color='green')
plt.title('Graph of sin(x)')
plt.xlabel('x values')
plt.ylabel('y = sin(x)')
plt.grid(True)
plt.legend()

# Adjust layout to prevent overlap between plots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame for demonstration
data = {'salary': [45000, 56000, 61000, 52000, 59000, 47000],
        'service': [5, 8, 10, 3, 6, 9]}
df = pd.DataFrame(data)

# Create subplots: 2 rows and 2 columns
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# First subplot: Line plot of x and y
x = range(6)
y = df['salary']
ax[0, 0].plot(x, y, marker='o', linestyle='-', color='blue', label='Salary Line')
ax[0, 0].set_title('Line Plot of Salary')
ax[0, 0].set_xlabel('Index')
ax[0, 0].set_ylabel('Salary')
ax[0, 0].legend()
ax[0, 0].grid(True)

# Second subplot: Histogram of 'salary' column
ax[0, 1].hist(df['salary'], bins=5, color='green', edgecolor='black', alpha=0.7)
ax[0, 1].set_title('Histogram of Salary')
ax[0, 1].set_xlabel('Salary')
ax[0, 1].set_ylabel('Frequency')
ax[0, 1].grid(True)

# Third subplot: Boxplot of 'salary' column
ax[1, 0].boxplot(df['salary'], patch_artist=True, boxprops=dict(facecolor='orange', color='black'))
ax[1, 0].set_title('Boxplot of Salary')
ax[1, 0].set_ylabel('Salary')

# Fourth subplot: Histogram of 'service' column
ax[1, 1].hist(df['service'], bins=5, color='purple', edgecolor='black', alpha=0.7)
ax[1, 1].set_title('Histogram of Service Years')
ax[1, 1].set_xlabel('Service Years')
ax[1, 1].set_ylabel('Frequency')
ax[1, 1].grid(True)

# Automatically adjust subplot layout to fit labels and titles
plt.tight_layout()

# Show the final plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6]
y = [100, 200, 300, 400, 500, 600]

# Create a figure with 2x2 grid of subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# First subplot: Line plot with red dashed line
ax[0, 0].plot(x, y, color='red', linestyle='dashed', label='Line plot')
ax[0, 0].set_title('Line Plot')
ax[0, 0].set_xlabel('X Axis')
ax[0, 0].set_ylabel('Y Axis')
ax[0, 0].legend()
ax[0, 0].grid(True)

# Second subplot: Scatter plot
ax[1, 0].scatter(x, y, color='blue', label='Scatter plot')
ax[1, 0].set_title('Scatter Plot')
ax[1, 0].set_xlabel('X Axis')
ax[1, 0].set_ylabel('Y Axis')
ax[1, 0].legend()
ax[1, 0].grid(True)

# Third subplot: Pie chart
ax[0, 1].set_title('Pie Chart')
ax[0, 1].pie(y, labels=x, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'cyan', 'pink'])

# Fourth subplot: Histogram
ax[1, 1].hist(y, color='purple', edgecolor='black', bins=5)
ax[1, 1].set_title('Histogram of Y values')
ax[1, 1].set_xlabel('Value bins')
ax[1, 1].set_ylabel('Frequency')

# Adjust the layout to prevent overlapping elements
plt.tight_layout()

# Show the plot
plt.show()
```
